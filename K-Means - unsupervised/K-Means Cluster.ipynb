```python
# Perform the necessary imports
import numpy as np
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score

# Read data (no headers)
df = pd.read_csv("fish.csv", header=None)

# Split data
species = df[0]  # First column is species names
# row: select all rows  
# column: select all columns except 1st columns (which contains name). 
samples = df.iloc[:, 1:]  # [row, column]

# Since therer's only 4 labels in .csv file, so we go with k = 3 (clusters)
kmeans = KMeans(n_clusters = 4)

# Pipeline: it scales first, then do kmeans 
pipeline = make_pipeline(StandardScaler(), kmeans)

# Fit the pipeline to samples
pipeline.fit(samples)

# Calculate the cluster labels: labels
labels = pipeline.predict(samples)
# output: [1 0 0 0  ... 1 1 0 3 2]

df = pd.DataFrame({'labels':labels,'species':species})

# Create crosstab: ct
ct = pd.crosstab(df['labels'], df['species'])

# Display ct
print(ct)

# Compare cluster labels vs true species
ari = adjusted_rand_score(species, labels)
print(f"Adjusted Rand Index: {ari:.3f}")
# ARI ranges: 1.0 = perfect, 0 = random, negative = worse than random
```
