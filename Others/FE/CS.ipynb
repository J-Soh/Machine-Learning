- Binarizing Columns 
    - Turning a column into Yes/No or 0/1 based on a threshold.

```python
# EXPLANATION:
# Original Age column
Age: [25, 30, 15, 40, 18]

# Binarize: Is age >= 18? (Adult or not)
Age_binary: [1, 1, 0, 1, 1]  # 1 = Adult, 0 = Not adult

# Explanation:
# 25 → 1 (yes, adult)
# 30 → 1 (yes, adult)
# 15 → 0 (no, not adult)
# 40 → 1 (yes, adult)
# 18 → 1 (yes, adult)
```

```python
# 3 METHODS:
import pandas as pd

# Method 1: Using threshold
df['Adult'] = (df['Age'] >= 18).astype(int)
# Result: True → 1, False → 0

# Method 2: Using sklearn
from sklearn.preprocessing import Binarizer
binarizer = Binarizer(threshold=18)
df['Adult'] = binarizer.transform(df[['Age']])

# Method 3: Using numpy
df['Adult'] = np.where(df['Age'] >= 18, 1, 0)
```
